### DMN-Parcellation: Replicating Functional Preference Profiles

In this notebook, our aim is to re-generate the plots depicting functional preference profiles as bar plots as opposed to polar plots. The data is derived from 5 masks: the DMN as a whole, and then four sub-regions: the MFC, PCC, and bilateral TPJ. The data is derived from the original DMN-parcellation submission, which can be found <a href="https://github.com/DVS-Lab/dmn-parcellation/blob/master/Functional%20preference%20profiles.ipynb">here</a>.

#### An overview of our goals:
1. Re-generate the feature_importances.csv dataset for each mask: the DMN.nii, MFC.nii, PCC.nii, rTPJ.nii and the lTPJ.nii.
2. Within each dataset, we rename three terms: 'food' is changed to 'eating-disorder', 'disorder' is changed to 'schizophrenia' and 'disorder2' is changed to 'Alzheimer-Parkinson'.
3. Next, the following steps will be performed for each region separately. First, we import .csv files that contain the importances for each region, for each psychological or health topic, as well as their respective confidence interval range. Then, we subset this file out to create two disctinct dataframes: one that is specific to the cognitive features that we're interested in, and one for the disorder topics we investigate. 
4. Afterwards, we simply generate bar plots for each dataframe.

## Functional Preference Profiles for the Whole DMN

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#First, we import the dataset from our first region to process: the DMN

data_dmn = pd.read_csv('/Users/lindseytepfer/Desktop/CurrentProjects/0_REVISION_DMN/Analyses/final-output/dmn-data.csv')
```

```python
cogdmn=data_dmn[data_dmn['feature'].isin(['memory','categorization','switching','inhibition',
                             'priming','mentalizing','fear','emotion','learning','reward',
                             'decision-making','imagery','spatial','attention','WM',
                             'awareness','language','arithmetic','semantics','reading'])]
```

```python
sns.set(rc={'axes.facecolor':'#f2f2f2', 'figure.facecolor':'white'})
fig, ax = plt.subplots()

cogdmn[cogdmn['region'] == 1].plot(x='feature', y='importance', ax=ax, position=.8,figsize=(20,15),
                                   kind='bar', title='DMN Functional Preference Profiles: Cognition',fontsize=25,
                                   label='Region 1', yerr='ci_range', width=.2, capsize=4, color='purple',
                                   alpha=0.6)

cogdmn[cogdmn['region'] == 2].plot(x='feature', y='importance', ax=ax, position=2, width=.15,
                                   kind='bar', label='Region 2',
                                   yerr='ci_range', capsize=4, color='green', alpha=0.6)

cogdmn[cogdmn['region'] == 3].plot(x='feature', y='importance', ax=ax, position=3, width=.15,
                                   kind='bar', label='Region 3', yerr='ci_range',
                                   capsize=4, color='red', alpha=0.6)

cogdmn[cogdmn['region'] == 4].plot(x='feature', y='importance', ax=ax,position=4,width=.15,
                                   kind='bar', label='Region 4', yerr='ci_range',
                                   capsize=4, color='blue', alpha=0.6)

plt.axhline(0, color='black')
ax.legend(fontsize=22,loc='lower right')
plt.ylabel('DMN Region Cognitive Importances', fontsize=42)
plt.xlabel('Features', fontsize=42)
ax.set_ylim(-1.8,1.8)

ax.tick_params(labelsize=22)
ax.title.set_size(25)



plt.xticks(rotation=70)
plt.tight_layout()

#plt.savefig("/Users/lindseytepfer/Desktop/CurrentProjects/0_REVISION_DMN/Analyses/final-output/dmn_cog_sns.svg", format="svg")
```
