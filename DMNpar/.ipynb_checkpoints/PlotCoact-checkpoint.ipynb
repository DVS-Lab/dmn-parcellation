```python
%matplotlib inline
import numpy as np
import seaborn as sns
from neurosynth.base.dataset import Dataset
dataset = Dataset.load("data/neurosynth_60_0.6.pkl")
from coactivation import coactivation_contrast
from plotting import make_thresholded_slices
```

```python
cut_coords = np.arange(-12, 50, 8) 
```

```python
contrast_mas = coactivation_contrast(dataset, 'masks/newDMN_4regions.nii.gz', q = 0.01)
fig_z = make_thresholded_slices(contrast_mas, sns.color_palette('Set1', 4), 
                       cut_coords=cut_coords, annotate=True)

fig_y = make_thresholded_slices(contrast_mas, sns.color_palette('Set1', 4), 
                       cut_coords=range(-60, 50, 18), display_mode='y', annotate=True)

fig_z.savefig('coactivation/DMN_coact_z.svg')
fig_y.savefig('coactivation/DMN_coact_y.svg')
```

```python
colors = ["amber","dark blue grey"]
colors = sns.xkcd_palette(colors)
contrast_mas = coactivation_contrast(dataset, 'images/MPFC/cluster_labels_k2.nii.gz', q = 0.01)
fig_z = make_thresholded_slices(contrast_mas, colors, 
        cut_coords=np.arange(-20, 20, 5), annotate=True)

fig_y = make_thresholded_slices(contrast_mas, sns.color_palette(colors), 
        cut_coords=range(0, 0, 4), display_mode='y', annotate=True)

fig_z.savefig('coactivation/MPFC_coact_z.svg')
fig_y.savefig('coactivation/MPFC_coact_y.svg')
```
